```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
from sklearn.decomposition import PCA
# init_notebook_mode(connected = True)
```

```python
# Загрузим данные
df = pd.read_csv("./behavior.csv", index_col=['Unnamed: 0'], parse_dates=['Unnamed: 0'])
```

```python
# Избавляемся от пустых строк
df = df.dropna(how='all')
```

```python
df[df['tag']==0].shape
```

```python
df[df['tag']==1].shape
```

Классы уравновешены

```python
# Отделяем ответы
X = df.drop('tag', axis=1)
y = df['tag']
```

```python
X.head()
```

```python
# Посмотрим матрицу корреляций
# corrmat = df.corr()
# top_corr_features = corrmat.index
# plt.figure(figsize=(20,20))
# g = sns.heatmap(df[top_corr_features].corr(),annot=True,cmap="RdYlGn")
df.corr()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Делим на тест и трейн
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
# Используем метод главных компонент, чтобы уменьшить количество фичей
pca = PCA(.95)
```

```python
pca.fit(X_train)
```

```python
# Число компонент
pca.n_components_
```

```python
X_train = pca.transform(X_train)
X_test = pca.transform(X_test)
```

### Логистическая регрессия

```python
from sklearn.linear_model import LogisticRegression
```

```python
logisticRegr = LogisticRegression(random_state=42, solver = 'lbfgs')
```

```python
logisticRegr.fit(X_train, y_train)
```

```python
logisticRegr.predict(X_test)
```

```python
score = logisticRegr.score(X_test, y_test)
print(score)
```

### Дерево решений

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
regressor = RandomForestClassifier(n_estimators=20, random_state=0)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```
